# .melt()
Aqui temos a recriação de uma porção dos dados originais.

```python
data = pd.DataFrame({'Product_Code': ['P1','P2','P3'],
                               'W0': [11, 7, 7],
                               'W1': [12, 6, 11],
                               'W2': [10, 3, 8]})
data
```

**id_vars** refere-se a coluna(s) que terá seus valores repetidos e guiará as alterações das outras.<br> 
**var_name** será o nome da nova coluna com as separações das colunas antigas.<br>
**value_name**  será os valores das colunas antigas (colunas que agora estão lado a lado com esses valores). 


      [var_name]    [value_name]
     Nova Coluna   Coluna com valores
      
    (coluna antiga)   (valor)
          W0            11
          W0             7
         ...           ...
          W1            12         

```python
df = data.melt(id_vars='Product_Code', var_name='Week', value_name='Sales')
df
```

**Nesse próximo passo, devemos lembrar que o algoritmo de Machine learning faz vários cálculos, por isso devemos transformar os dados categóricos que em números (há métodos machine learnincos para isso, mas usar os métodos de strings(str) do próprio python já resolve nosso problema).**

```python
# Em outras palavras: Mantendo apenas os str(números) e convertendo-os para int(números) inteiros (para fazermos os cálculos)
df['Product_Code'] = df['Product_Code'].str.extract('(\d+)', expand=False).astype(int)
df
```

```python
df['Week'] = df['Week'].str.extract('(\d+)', expand=False).astype(int)
df
```

```python
# Ordena os valores, se bem que nesse dataset eles já estão ordenados (mas não temos como ter certeza em um com milhares de
# linhas, na dúvida, ordene)
df = df.sort_values(['Week', 'Product_Code'])
df
```
