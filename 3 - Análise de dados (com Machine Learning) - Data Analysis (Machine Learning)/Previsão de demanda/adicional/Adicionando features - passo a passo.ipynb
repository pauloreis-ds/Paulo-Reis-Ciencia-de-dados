**.diff()** calcula a diferença entre o valor de uma linha e o valor da linha anterior

```python
hehe = pd.DataFrame({'Agrupar':[1,2,3,4,5],
                      'Testar':[2,4,8,0,0]})
hehe
```

(linha 0 - não há linha = NaN) 1-?? <br>
(linha 1 - linha 0) ---------- 2-1 <br>
(linha 2 - linha 1) ---------- 3-2 <br>
(linha 3 - linha 2) ---------- 4-3 <br>
(linha 4 - linha 3) ---------- 5-4 <br>

```python
hehe.diff()
```

**.groupby()** simplesmente agrupa os dados de acordo com os valores únicos da coluna.

         Mais ou menos assim:
    Normal               Agrupado
    1    a              1    [a,D]
    2    b              2    [b,E]
    3    c              3    [c,F]
    1    D
    2    E
    3    F

```python
haha = pd.DataFrame({'Agrupar':[1,2,3,1,2,3],
                      'Testar':['a','b','c','d','e','descerei'],
                'Testar_shift':[10,11,12,20,21,22]})
haha
```

```python
haha.groupby('Agrupar').count()
```

No nosso mini dataset o resultado é:

```python
df2 = df.copy()
df2.groupby(['Product_Code']).count()
```

## .shift()
Como você pode ver, ele desloca os valores para a linha abaixo.

```python
haha
```

```python
haha.shift()
```

```python
haha.groupby('Agrupar').count()
```

```python
# O ponto é...
haha.groupby('Agrupar').shift()
```

```python
#... faremos isso apenas em uma coluna, não em todo o dataset
haha['Coluna_shift'] = haha.groupby('Agrupar')['Testar_shift'].shift()
haha
```

No nosso dataset:

```python
df2 = df.copy()
df2['Last_Week_Sales'] = df2.groupby(['Product_Code'])['Sales'].shift()
df2
```

O mesmo mas aplicando **.diff()** para ter o resultado desejado.

```python
df2['Last_Week_Diff'] = df2.groupby(['Product_Code'])['Last_Week_Sales'].diff()
df2
```

E agora eliminando as linhas com os valores nulos (NaN), podemos utilizar esses dados para validar nosso modelo através de uma **baseline**.

A baseline será o resultado de um modelo/uma solução muito básica.

Geralmente, criamos uma baseline e então tentamos criar soluções mais complexas para obter um melhor resultado. Se você obtiver uma pontuação melhor que a baseline, é um bom sinal.
